```python
import pandas as pd
import pickle
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from scipy.sparse import save_npz, vstack

# Define file path
file_path = r"C:/Users/dell/Documents/Guvi Lectures/Swiggy model/swiggy_cleaned.csv"

# Load a sample to inspect the column types
df_sample = pd.read_csv(file_path, nrows=50_000)

# Convert '--' to NaN
df_sample.replace('--', np.nan, inplace=True)

# Identify numeric columns properly
numeric_cols = ["cost", "rating"]  # Ensure these are the correct numeric columns

# Convert only numeric columns safely
for col in numeric_cols:
    df_sample[col] = pd.to_numeric(df_sample[col], errors="coerce")  # Coerce invalid values to NaN

# Fill missing numeric values with median
df_sample[numeric_cols] = df_sample[numeric_cols].fillna(df_sample[numeric_cols].median())

# Check all categorical columns before encoding
categorical_cols = ["name", "city", "cuisine"]  
df_sample[categorical_cols] = df_sample[categorical_cols].astype(str)  # Convert categorical columns to string

# Apply Label Encoding for 'name' and 'city'
le_name = LabelEncoder()
le_city = LabelEncoder()
df_sample["name"] = le_name.fit_transform(df_sample["name"])
df_sample["city"] = le_city.fit_transform(df_sample["city"])

# Apply One-Hot Encoding for 'cuisine'
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
encoder.fit(df_sample[["cuisine"]])

# Initialize an empty list for sparse matrices
sparse_matrices = []

# Process dataset in chunks
chunksize = 10_000  
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunksize)):
    print(f"Processing chunk {i+1}...")

    # Convert '--' to NaN
    chunk.replace('--', np.nan, inplace=True)

    # Convert only numeric columns safely
    for col in numeric_cols:
        chunk[col] = pd.to_numeric(chunk[col], errors="coerce")

    # Fill missing values with median
    chunk[numeric_cols] = chunk[numeric_cols].fillna(df_sample[numeric_cols].median())

    # Convert categorical columns to string before encoding
    chunk[categorical_cols] = chunk[categorical_cols].astype(str)

    # Apply Label Encoding using previously fitted encoders
    chunk["name"] = le_name.transform(chunk["name"])
    chunk["city"] = le_city.transform(chunk["city"])

    # Apply One-Hot Encoding for 'cuisine'
    encoded_chunk = encoder.transform(chunk[["cuisine"]])

    # Convert numeric data to float32 to save memory
    chunk = chunk.drop(columns=["cuisine"]).astype("float32")

    # Store sparse matrix
    sparse_matrices.append(encoded_chunk)

    # Save processed chunk
    chunk.to_csv(f"C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoded_chunk_{i}.csv", index=False)

# Stack all sparse chunks into one matrix
final_sparse_matrix = vstack(sparse_matrices)

# Save final sparse matrix
save_npz("/mnt/data/encoded_data.npz", final_sparse_matrix)

# Save encoders
with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoder.pkl", "wb") as file:
    pickle.dump(encoder, file)

with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/le_name.pkl", "wb") as file:
    pickle.dump(le_name, file)

with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/le_city.pkl", "wb") as file:
    pickle.dump(le_city, file)

print("✅ Data processing completed! Encoded data saved as sparse matrix in encoded_data.npz")
```

```python
import pandas as pd
import pickle
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from scipy.sparse import save_npz, vstack

# Define file path
file_path = "C:/Users/dell/Documents/Guvi Lectures/Swiggy model/swiggy_cleaned.csv"

# Load a sample to inspect the column types
df_sample = pd.read_csv(file_path, nrows=50_000)

# Drop 'address' column (not needed for encoding)
df_sample.drop(columns=["address"], inplace=True)

# Convert '--' to NaN
df_sample.replace('--', np.nan, inplace=True)

# Identify numeric columns
numeric_cols = ["cost", "rating"]

# Convert only numeric columns safely
for col in numeric_cols:
    df_sample[col] = pd.to_numeric(df_sample[col], errors="coerce")  # Coerce invalid values to NaN

# Fill missing numeric values with median
df_sample[numeric_cols] = df_sample[numeric_cols].fillna(df_sample[numeric_cols].median())

# Check categorical columns before encoding
categorical_cols = ["name", "city", "cuisine"]
df_sample[categorical_cols] = df_sample[categorical_cols].astype(str)  # Convert categorical columns to string

# Apply Label Encoding for 'name' and 'city'
le_name = LabelEncoder()
le_city = LabelEncoder()
df_sample["name"] = le_name.fit_transform(df_sample["name"])
df_sample["city"] = le_city.fit_transform(df_sample["city"])

# Apply One-Hot Encoding for 'cuisine'
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
encoder.fit(df_sample[["cuisine"]])

# Save the LabelEncoders for later use
with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/le_name.pkl", "wb") as file:
    pickle.dump(le_name, file)

with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/le_city.pkl", "wb") as file:
    pickle.dump(le_city, file)

# Initialize an empty list for sparse matrices
sparse_matrices = []

# Process dataset in chunks
chunksize = 10_000  
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunksize)):
    print(f"Processing chunk {i+1}...")

    # Drop 'address' column
    chunk.drop(columns=["address"], inplace=True)

    # Convert '--' to NaN
    chunk.replace('--', np.nan, inplace=True)

    # Convert only numeric columns safely
    for col in numeric_cols:
        chunk[col] = pd.to_numeric(chunk[col], errors="coerce")

    # Fill missing values with median
    chunk[numeric_cols] = chunk[numeric_cols].fillna(df_sample[numeric_cols].median())

    # Convert categorical columns to string before encoding
    chunk[categorical_cols] = chunk[categorical_cols].astype(str)

    # Handle unseen labels in 'name' and 'city' using mapping
    chunk["name"] = chunk["name"].map(lambda x: le_name.transform([x])[0] if x in le_name.classes_ else -1)
    chunk["city"] = chunk["city"].map(lambda x: le_city.transform([x])[0] if x in le_city.classes_ else -1)

    # Apply One-Hot Encoding for 'cuisine'
    encoded_chunk = encoder.transform(chunk[["cuisine"]])

    # Convert numeric data to float32 to save memory
    chunk = chunk.drop(columns=["cuisine"]).astype("float32")

    # Store sparse matrix
    sparse_matrices.append(encoded_chunk)

    # Save processed chunk
    chunk.to_csv(f"C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoded_chunk_{i}.csv", index=False)

# Stack all sparse chunks into one matrix
final_sparse_matrix = vstack(sparse_matrices)

# Save final sparse matrix
save_npz("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoded_data.npz", final_sparse_matrix)

# Save the One-Hot Encoder
with open("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoder.pkl", "wb") as file:
    pickle.dump(encoder, file)

print("✅ Data processing completed! Encoded data saved as sparse matrix in encoded_data.npz")
```

```python
import numpy as np
from scipy.sparse import load_npz
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import TruncatedSVD

# Load sparse encoded dataset (Do NOT convert to dense)
encoded_data = load_npz("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoded_data.npz")

# Reduce dimensions using Truncated SVD (works with sparse matrices)
svd = TruncatedSVD(n_components=50, random_state=42)
encoded_data_reduced = svd.fit_transform(encoded_data)

# Choose optimal number of clusters
k = 10  # Can be tuned using the Elbow method

# Train MiniBatchKMeans (Efficient for large datasets)
kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=1000)
clusters = kmeans.fit_predict(encoded_data_reduced)

# Assign cluster labels to restaurants
restaurant_clusters = {i: cluster for i, cluster in enumerate(clusters)}

# Function to get recommendations from the same cluster
def recommend_restaurants(index, top_n=5):
    cluster = restaurant_clusters.get(index)  # Find the cluster for the given restaurant
    if cluster is None:
        return []  # Return empty if index is invalid
    similar_indexes = [i for i, c in restaurant_clusters.items() if c == cluster and i != index]
    return similar_indexes[:top_n]

# Example usage: Get 5 recommendations for restaurant at index 10
recommended_indexes = recommend_restaurants(index=10, top_n=5)
print("Recommended restaurant indices:", recommended_indexes)
```

```python
import numpy as np
from scipy.sparse import load_npz
from sklearn.metrics.pairwise import cosine_similarity

# Load the sparse encoded dataset (DO NOT convert to dense)
encoded_data = load_npz("C:/Users/dell/Documents/Guvi Lectures/Swiggy model/encoded_data.npz")

# Function to get top N similar restaurants (efficient memory usage)
def get_similar_restaurants(index, top_n=5):
    # Compute similarity ONLY for the given restaurant
    similarity_scores = cosine_similarity(encoded_data[index], encoded_data).flatten()

    # Get top N most similar restaurant indices (excluding itself)
    similar_indices = np.argsort(similarity_scores)[::-1][1:top_n+1]

    return similar_indices

# Example usage: Get 5 most similar restaurants for index 10
similar_restaurants = get_similar_restaurants(index=10, top_n=5)
print("Recommended restaurant indices:", similar_restaurants)
```
